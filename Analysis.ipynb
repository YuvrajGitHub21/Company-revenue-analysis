```python
df['Date'] = pd.to_datetime(df['Date'], format='%m/%d/%Y')

# Set 'Date' column as the index
df.set_index('Date', inplace=True)
```

```python
# df.set_index('Date', inplace=True)

monthly_sales = df['Sales'].resample('ME').sum()
quarterly_sales = df['Sales'].resample('QE').sum()

# Create line graphs for monthly and quarterly sales
plt.figure(figsize=(14, 5))

# Monthly line graph
plt.subplot(1, 2, 1)
plt.plot(monthly_sales, marker='o')
plt.title('Monthly Sales')

# Quarterly line graph
plt.subplot(1, 2, 2)
plt.plot(quarterly_sales, marker='o')
plt.title('Quarterly Sales')

# Show the plots
plt.tight_layout()
plt.show()
```

```python
# df.set_index('Date', inplace=True)

monthly_profit = df['Profit'].resample('ME').sum()
quarterly_profit = df['Profit'].resample('QE').sum()

# Create line graphs for monthly and quarterly sales
plt.figure(figsize=(14, 5))

# Monthly line graph
plt.subplot(1, 2, 1)
plt.plot(monthly_profit, marker='o')
plt.title('Monthly Profit')

# Quarterly line graph
plt.subplot(1, 2, 2)
plt.plot(quarterly_profit, marker='o')
plt.title('Quarterly Profit')

# Show the plots
plt.tight_layout()
plt.show()
```
